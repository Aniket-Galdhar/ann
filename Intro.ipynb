```python
import numpy as np
import matplotlib.pyplot as plt
```

### Binary Activation Function

```python
def binary_activation(x):
    """Returns 0 if x <= 0, or 1 if x > 1"""
    return np.heaviside(x,1)
```

```python
x = np.arange(-10, 10, 0.1)
plt.figure(figsize=(8,2.5))
plt.plot(x, binary_activation(x))
plt.show()
```

### Linear Activation function

```python
def linear_function(x):
    return x
```

```python
x = np.arange(-10, 10, 0.1)
plt.figure(figsize=(8,2.5))
plt.plot(x, linear_function(x))
plt.show()
```

### sigmoid

```python
def sigmoid(x):
    # return 1/(1+np.exp(-x))
    x1=[]
    for i in x:
        if i<=0:
            x1.append(0)
        else:
            x1.append(1)

    return x1
```

```python
x = np.arange(-10, 10, 0.1)
plt.figure(figsize=(8,2.5))
plt.plot(x, sigmoid(x))
plt.show()
```

### tanh

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.arange(-10, 10, 0.1)
plt.figure(figsize=(8,2.5))
plt.plot(x, tanh(x))
plt.show()
```

### Relu

```python
def relu(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.arange(-10, 10, 0.1)
plt.figure(figsize=(8,2.5))
plt.plot(x, relu(x))
plt.show()
```

### Softmax

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.arange(-10, 10, 0.1)
plt.figure(figsize=(8,2.5))
plt.plot(x, softmax(x))
plt.show()
```
